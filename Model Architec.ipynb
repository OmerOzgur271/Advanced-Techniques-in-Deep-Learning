```python
from keras.models import Model
from keras.models import Sequential
from keras.layers import Dense,Input,Concatenate,concatenate
from keras.utils.vis_utils import plot_model
from keras.layers import Input, Add, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D, AveragePooling2D, MaxPooling2D, GlobalMaxPooling2D,ReLU
```

```python
input1 = Input(shape=(10,),name="Giris")
x = Dense(32, activation='relu',name="Gizli")(input1)
#x = Dense(20, activation='relu')(x)
#x = Dense(10, activation='relu')(x)
output = Dense(1, activation='sigmoid',name="Cikti")(x)
model = Model(inputs=input1, outputs=output)

plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
input1 = Input(shape=(10,),name="Giris")
l1 = Dense(10, activation='relu',name="Gizli")(input1)
l2 = Dense(10, activation='relu')(input1)
l3 = Dense(10, activation='relu')(l1)
l4 = Dense(10, activation='relu')(l2)
l5 = Dense(10, activation='relu')(input1)
merge = concatenate([l3,l4,l5])
#x = Dense(10, activation='relu')(x)
output = Dense(1, activation='sigmoid',name="Cikti")(merge)

model = Model(inputs=input1, outputs=output)

plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
input1 = Input(shape=(16,16,16))
x1 = Conv2D(8,kernel_size=(3,3),activation='relu',padding="same")(input1)
x2 = Conv2D(16,kernel_size=(3,3),activation='relu',padding="same")(x1)
x3 = Conv2D(16,kernel_size=(3,3),activation='relu',padding="same")(x2)
added = Add()([input1, x3])

out = ReLU()(added)
model = Model(inputs=[input1, input2], outputs=out)
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```
